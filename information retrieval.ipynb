```python
#importing libraries
import pandas as pd
import nltk
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer , TfidfVectorizer
from sklearn.neighbors import KDTree
```

```python
person=pd.read_csv('famous_people.csv')
```

```python
person.head(5)
```

```python
from nltk.corpus import stopwords
import re
from nltk.stem import WordNetLemmatizer
lemma=WordNetLemmatizer()
```

```python
person['Text']=person['Text'].apply(lambda x: x.lower())
person['Text']=person['Text'].apply(lambda x : re.sub(r"@\S+" , "" , x))
person['Text']=person['Text'].apply(lambda x : lemma.lemmatize(x,pos='v'))

stop=stopwords.words('english')

person['Text']=person['Text'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop]))
```

```python
person
```

```python
person.shape
```

```python
tfidf=TfidfVectorizer()
train_tfidf=tfidf.fit_transform(person.Text)
train_tfidf=train_tfidf.toarray()
```

```python
person['tfidf']=list(train_tfidf)
```

```python
person
```

```python
# training on the kdtree
kdtree = KDTree(train_tfidf)
```

```python
distance  , idx=kdtree.query(person.tfidf[16].reshape(1,-1) , k = 4 )
```

```python
idx,distance
```

```python
for i , value in list(enumerate(idx[0])):
    print(f'Name : {person["Name"][value]}')
    print(f'Distance : {distance[0][i]}')    
    print(f'URI : {person["URI"][value]}')    
```

```python
# saving of a model
import joblib
#save the tfidf vectorizer
joblib.dump(tfidf , 'tfidf_vector_model.pkl')


# saving the kdtree model
joblib.dump(kdtree , 'kdtree_model.pkl')
```
